##### Temos quantos funcionários do sexo masculino e feminino na nossa base? E qual o gasto salarial dos mesmo?

```python
# Quantidade de funcionários por gênero
total_p_genero = (
    df['Genero']
    .value_counts()
    .reset_index(name='qtd pessoas')
)
print(total_p_genero)


# Gasto salarial po gênero
media_salarial = (
    df.groupby('Genero')['Salario']
    .sum()
    .round(2)
    .reset_index(name='Gasto Salarial')
)
print(media_salarial)


#df.head()
```

##### Temos quantos funcionários em cada cidade? E quais são essas cidades?

```python
# Quantidade de funcionários em cada cidade
pessoas_p_cidade = (
    df['Cidade']
    .value_counts()
    .reset_index(name='qtd_pessoas')
    .sort_values(by='qtd_pessoas', ascending=False)
)
print(pessoas_p_cidade)


# E quais são as cidades que os nossos funcionários residem?
print('\nAs cidades são:')
cidades = (
    df['Cidade']
    .unique()
)
print(cidades)



df.head()
```

##### Temos quantos Diretores, Gerentes, Coordenadores, Analistas e Estagiários na nossa base? E qual o gasto salarial completo por cada cargo?

```python

# Quantidade de pessoas por cargo
qtd_p_cargo = (
    df['Cargo']
    .value_counts()
    .reset_index(name='Qtd Funcionários')
)
print(qtd_p_cargo)


# Gasto salarial por cargos
print('\n')
gasto_salarial_p_cargo = (
    df.groupby('Cargo')['Salario']
    .sum()
    .reset_index(name='Gastos Salariais')
    .sort_values(by='Gastos Salariais', ascending=False)
)
print(gasto_salarial_p_cargo)
```

##### Quanto que a empresa gasta de Salario, Vale Refeição(VR) e Vale Transporte(VT)?

```python
# Gastos com Salario
salario = (
    df['Salario']
    .sum()
)
print(f'A empresa gasta um total de R${salario} com salário.')


# Gastos com Vale Refeição(VR)
print('\n')
vr = (
    df['VR']
    .sum()
)
print(f'A empresa gasta um total de R${vr} com Vale Refeição(VR).')


# Gastos com Vale Transporte(VT)
print('\n')
vt = (
    df['VT']
    .sum()
)
print(f'A empresa gasta um total de R${vt} com Vale Transporte(VT)')

```

### Agora, crie um indicador que mostre o nível performático do funcionário<br>
##### Para isso, siga o modelo abaixo:<br>
##### 9-10 -> Excelente
##### 8-8.9 -> Muito Bom
##### 7-7.9 -> Bom
##### 5-6.9 -> Regular
##### <5 -> Ruim

```python
def avaliacao_funcionario(df):
    if df > 9:
        return 'Excelente'
    elif df > 8:
        return 'Muito Bom'
    elif df > 7:
        return 'Bom'
    elif df > 5:
        return 'Regular'
    else:
        return 'Ruim'

df['Classificacao por Desempenho'] = df['Avaliação do Funcionário'].apply(avaliacao_funcionario)

df.head()
```

##### Existe algum funcionário com classificação RUIM?

```python
df[df['Classificacao por Desempenho'] == 'Ruim']

## Não existe nenhum funcionário com a classificação "Ruim"!
```

##### Qual a  quantidade de funcionários pela Classificação por Desempenho? 

```python
classificacao_desempenho = (
    df.groupby('Classificacao por Desempenho')['ID RH']
    .count()
    .reset_index(name='Qtd de Pessoas')
    .sort_values(by='Qtd de Pessoas', ascending=False)
)
print(classificacao_desempenho)
```

```python
classificacao_desempenho_distribuicao = (
    df.groupby('Classificacao por Desempenho')['Classificacao por Desempenho']
    .count()
)
#print(classificacao_desempenho_distribuicao)

plt.figure(figsize=(10, 8))

classificacao_desempenho_distribuicao.plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'edgecolor':'black'},
    shadow=True
)

plt.title('Distribuiçao pela Classificação por Desempenho', fontsize=14, fontweight='bold')
plt.legend(classificacao_desempenho_distribuicao.index, title='Legenda', loc='upper right')

plt.show()
```

##### "47,4% dos funcionários (111 pessoas) estão na categoria Muito Bom, representando a principal concentração de desempenho na empresa."
